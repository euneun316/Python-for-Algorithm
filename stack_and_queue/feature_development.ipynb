```python
from collections import deque
from math import ceil

def solution(progresses, speeds):
    answer = []
    # ì†Œìš”ì¼ ë¦¬ìŠ¤íŠ¸í™”
    spend_date = [ceil((100-p)/s) for p, s in zip(progresses, speeds)]
    
    while spend_date:
        cnt = 1
        pivot = spend_date.pop(0)           # ê¸°ì¤€ì¼ ì„¤ì •
        for date in spend_date:
            if pivot >= date:               # ê¸°ì¤€ì¼ì´ ë‹¤ìŒë‚ ë³´ë‹¤ í¬ë‹¤ë©´
                cnt += 1                    # ì¹´ìš´íŠ¸ 1(í•¨ê»˜ ë°°í¬)
                spend_date = spend_date[1:] # ë‹¤ë‹¤ìŒë‚ ë¡œ ì´ë™
            else:
                break
        answer.append(cnt)                  # ì¹´ìš´íŠ¸ ì¶”ê°€
    
    return answer
```
