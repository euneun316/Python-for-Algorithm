ì´ ë¬¸ì œëŠ” ë¹„íŠ¸ ì—°ì‚°Bitwise Operationì„ ë¬»ëŠ” ë¬¸ì œì…ë‹ˆë‹¤. ì´ë¯¸ ë¬¸ì œ ì˜ˆì‹œì— 2ì§„ìˆ˜ë¡œ ì²˜ë¦¬í•˜ëŠ” íŒíŠ¸ê°€ í¬í•¨ë˜ì–´ ìˆê³ , ë‘˜ ì¤‘ í•˜ë‚˜ê°€ 1ì¼ ê²½ìš°ì— ë²½ `#`ì´ ìƒê¸°ê¸° ë•Œë¬¸ì— ORë¡œ ì²˜ë¦¬í•˜ë©´ ê°„ë‹¨íˆ í’€ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì•„ì£¼ ì‰¬ìš´ ë¬¸ì œì˜€ë˜ ë§Œí¼ `if else`ë¡œ í’€ì´í•œ ë¶„ë“¤ë„ ë§ì´ ë°œê²¬ë˜ì—ˆëŠ”ë°ìš”. ì •ë‹µìœ¼ë¡œëŠ” ê°„ì£¼ë˜ì§€ë§Œ ì´ ë¬¸ì œëŠ” ë¹„íŠ¸ ì—°ì‚°ì„ ì˜ ë‹¤ë£° ìˆ˜ ìˆëŠ”ì§€ë¥¼ ë¬»ê³ ì í•˜ëŠ” ì˜ë„ì˜€ë˜ ë§Œí¼ ì•ìœ¼ë¡œ ì´ëŸ° ìœ í˜•ì˜ ë¬¸ì œë¥¼ í’€ ë•ŒëŠ” ë¹„íŠ¸ ì—°ì‚°ì„ ê¼­ ê¸°ì–µí•˜ì‹œê¸° ë°”ëë‹ˆë‹¤.

```python
def solution(n, arr1, arr2):
    answer = []
    for i in range(n):
        bin_str = bin(arr1[i] | arr2[i])[2:] # 0bìë¦„
        # 0ìœ¼ë¡œ ì‹œì‘í•˜ëŠ” ê²½ìš° : "0" *(n - len(bin_str))
        answer.append(("0" *(n - len(bin_str)) + bin_str).replace("1", "#")
                                                      .replace("0", " "))
    return answer
```
