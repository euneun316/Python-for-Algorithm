## üîó Î¨∏Ï†ú ÎßÅÌÅ¨

$ H = \displaystyle\sum_{i=0}^{l-1}{a_ir^i}\; mod M$

````python
L = int(input().strip()) # Î¨∏ÏûêÏó¥ Í∏∏Ïù¥
M = 1234567891 
r = 31

_str = input().strip() # ÏÜåÎ¨∏Ïûê Î¨∏ÏûêÏó¥ ÏûÖÎ†•

answer = 0

for i in range(len(_str)):
    # ord : Return the Unicode code
    ai = ord(_str[i]) - ord('a') + 1 # a = 1
    answer += ai * (r ** i)
    # mod M ÎÇòÎ®∏ÏßÄ Í≥ÑÏÇ∞
print(answer % M)
````
