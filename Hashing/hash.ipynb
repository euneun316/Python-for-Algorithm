## 딕셔너리 루프

```python
hash = dict()
```

```python
for i in range(1,6):
    hash[i] = i ** 2
```

```python
hash
```

```python
for k in hash.keys():
    print(k)
```

```python
for k, v in hash.items():
    print(k,v)
```

## 딕셔너리 정렬
- sorted() : 언제나 list 타입을 반환

```python
hash = dict({1:10,3:12,5:7,7:6,4:5})
```

```python
# 오름차순 정렬 keys
sorted(hash.keys(), key=lambda x : x)
```

```python
# 오름차순 정렬 values
sorted(hash.values(), key=lambda x : x)
```

```python
# 오름차순 정렬 items - keys
sorted(hash.items(), key=lambda x : x[0])
```

```python
# 오름차순 정렬 items - values
sorted(hash.items(), key=lambda x : x[1])
```

```python
# 내림차순 정렬 keys
sorted(hash.keys(), key=lambda x : -x)
```

```python
# 내림차순 정렬 values
sorted(hash.values(), key=lambda x : -x)
```

```python
# 내림차순 정렬 key values - error
# tuple에 마이너스 줄 수 없기 때문
# sorted(hash.items(), key=lambda x : - x)

#내림차순 정렬 items - keys
sorted(hash.items(), key=lambda x : - x[0])
```

```python
#내림차순 정렬 items - values
sorted(hash.items(), key=lambda x : - x[1])
```
